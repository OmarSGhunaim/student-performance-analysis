```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("student-mat.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.barplot(x="studytime", y="G3", data=df)
plt.title("Study Time vs Final Grade")
plt.show()
```

```python
sns.boxplot(x="sex", y="G3", data=df)
plt.title("Gender vs Final Grade")
plt.show()
```

Students who study more tend to get higher grades

Grade distribution varies slightly between genders

Higher study time leads to more consistent results
